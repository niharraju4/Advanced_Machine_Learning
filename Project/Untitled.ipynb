```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_reviews = pd.read_csv('N:/advanced ML all files/Project/googleplaystore_user_reviews.csv')
df_apps = pd.read_csv('N:/advanced ML all files/Project/googleplaystore.csv')
```

```python
# Assuming 'App' is the common column
merged_df = pd.merge(df_reviews, df_apps, on='App', how='inner')
```

```python
merged_df.to_csv('merged_data.csv', index=False)
```

```python
# Load your dataset
df = pd.read_csv('merged_data.csv')

# Fill missing values
df['Sentiment_Polarity'].fillna(df['Sentiment_Polarity'].mean(), inplace=True)
df['Sentiment_Subjectivity'].fillna(df['Sentiment_Subjectivity'].mean(), inplace=True)
df['Rating'].fillna(df['Rating'].mode()[0], inplace=True)
df['Translated_Review'].fillna('Unknown', inplace=True)
df['Sentiment'].fillna('Unknown', inplace=True)

# Convert 'Reviews' to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# One-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['Category', 'Type', 'Content Rating', 'Genres'], drop_first=True)

# Now all your columns should be in a numeric format
df.info()
```
